```python
#Read Data from csv file
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv

# Load data from CSV file using numpy

Data = np.loadtxt(r"C:\Users\hp\Documents\mydata.csv",delimiter = ',' , skiprows = 1)
print(Data) 
```

```python
Newdata = Data.T
print(Newdata)
```

```python
# extrating data of Y1 
Y1 = Newdata[0]
print(Y1)
```

```python
# extrating data of Y2 
Y2 = Newdata[1]
print(Y2)
```

```python
# extrating data of X1 
X1 = Newdata[2]
print(X1)
```

```python
# extrating data of X1 
X2 = Newdata[3]
print(X2)
```

```python
# Reshape data for x1 and x2 and creating x0 vector having all 1's
n = len(X1)
x1 = np.array(X1).reshape(n,1)
x2 = np.array(X2).reshape(n,1)
y1 = np.array(Y1).reshape(n,1)
y2 = np.array(Y2).reshape(n,1)
x0 = np.ones([n,1])
print(" Y1 is  ")
print(" ")
print(y1)
print(" ")
print(" Y2 is  ")
print(" ")
print(y2)
print(" ")
print(" X1 is  ")
print(" ")
print(x1)
print(" ")
print(" X2 is ")
print(" ")
print(x2)
print(" ")
print(" x0 is ")
print(" ")
print(x0)
```

```python
# Normalize data for x1
m_x1 = np.mean(x1)
SD1 = np.std(x1)
norm_x1 = []
for xi in X1:
    norm_x1.append((xi-m_x1)/SD1)
print("Normalization of x1 is ")
print(" ")
print(norm_x1)
```

```python
# Normalize data for x2
m_x2 = np.mean(x2)
SD2 = np.std(x2)
norm_x2 = []
for xi in X2:
    norm_x2.append((xi-m_x2)/SD2)
print("Normalization of x2 is ")
print(" ")
print(norm_x2)
```

```python
# Now reshape norm_x1 and norm_x2 and converting x0 , norm_x1 and norm_x2 into a single matrix X
n = len(x1)
norm_X1 = np.array(norm_x1).reshape(n,1)
norm_X2 = np.array(norm_x2).reshape(n,1)
X0 = np.array(x0).reshape(n,1)
X = np.hstack((x0 , norm_X1 , norm_X2)) # hstack is a function in numpy used to concatenate x0 , norm_x1 and norm_x2
print("   X matrix is   ")
print(" ")
print(X)
```

```python
# Now reshape y1 and y2 , y1 and y2 into a single matrix Y
Y = np.hstack((y1 , y2)) # hstack is a function in numpy used to concatenate x0 , norm_x1 and norm_x2
print("   Y matrix is   ")
print(" ")
print(Y)
```

```python
# Taking transpose of matrix X
print("Transpose of X matrix is ")
print(" ")
trans_x = X.T
print(trans_x)
```

```python
# Correlation Matrix
cor =  trans_x.dot(X)/n
print(cor)
```

```python
# Now multiply trans_X and X matrix
print("  multiplication of X.T and X ")
print(" ")
trans_x_X =  trans_x.dot(X) # dot function is used for multiplication in numpy
print(trans_x_X)
```

**Correlation matrix is helpful in detecting problem of multicollinearity so in this dataset 
our daigonal matrix of correlation matrix are unity and off diagonal elements are not tends to unity and our
matrix is symmetric so there is no multicollinearity in our data**

```python
# Taking inverse of trans_x_X
print("  inverse of ((X.T).X) is ")
print(" ")
inv = np.linalg.inv(trans_x_X)
print(inv)
```

```python
# Now Multiply inverse of ((X.T).X) and X.T 
MUL = inv.dot(X.T)
print(MUL)
```

```python
# Now calculate beta 
print(" coefficient are ")
print(" ")
beta = MUL.dot(Y)
print(beta)
```

```python
# Predicted values of Y
Y_pred = X.dot(beta)
print(" Predicted values of Y are")
print(" ")
print(Y_pred)
```

```python
# Anova Table
m = np.shape(Y)[1]
n = len(X)
k = len(beta)

# Degree of freedom

# DF for Regression
DF_REG = m*(n-k)

# DF for Residual
DF_RES = m*(k-1)

# DF for Total
DF_TOT = m*(n-1)


# Sum of square
m_y = np.mean(Y)
error = Y - Y_pred

# SS Residual 
SS_res = np.sum(error**2)

# SS Regression

SS_Reg = np.sum((Y_pred - m_y)**2)

# SS Total
SS_total = np.sum((Y - m_y)**2)

# MS Residual

MS_Res = SS_res/DF_RES

# MS Regression

MS_Reg = SS_Reg/DF_REG

# F Statistics

F = MS_Reg/MS_Res

# Printing Anova table
print("Anova table ")
print(" ")
print('{0}\t\t{1}\t\t{2}\t\t\t{3}\t\t\t{4}'.format( "Sources","DF  " , "SS" , "MS" , "F statistics"))
print('{0}\t{1}\t{2}\t{3}\t\t{4}'.format( "Regression" ,DF_REG , SS_Reg , MS_Reg , F ))
print('{0}\t{1}\t{2}\t{3}'.format("Residual",DF_RES , SS_res , MS_Res))
print('{0}\t\t{1}\t{2}'.format("Total",  DF_TOT , SS_total ))
```
