```python
# Read data from csv
import matplotlib.pyplot as plt
import numpy as np
import csv

# Load data from CSV file using numpy

Data = np.loadtxt(r"C:\Users\hp\Documents\Linear DA.csv",delimiter = ',' , skiprows = 1)
print(Data)
```

```python
# Extract X1 , X2 and Class
C = Data[:len(Data),0:1]
print(" Class is ")
print(" ")
print(C)

print(" ")
print(" X1 is ")
print(" ")
X1 = Data[:len(Data),1:2]
print(X1)

print(" ")
print(" X2 is ")
print(" ")
X2 = Data[:len(Data),2:3]
print(X2)
```

```python
# Normalization of X1 
m_x1 = np.mean(X1)
SD1 = np.std(X1)
norm_x1 = []
for xi in X1:
    norm_x1.append((xi-m_x1)/SD1)
print("Normalization of x1 is ")
print(" ")
print(np.array(norm_x1))
```

```python
# Normaliztion of X2 is
m_x2 = np.mean(X2)
SD2 = np.std(X2)
norm_x2 = []
for xi in X2:
    norm_x2.append((xi-m_x2)/SD2)
print("Normalization of x2 is ")
print(" ")
print(np.array(norm_x2))
```

```python
# Now making X matrix
X = np.hstack((np.array(norm_x1) , np.array(norm_x2)))
print(X)
```

```python
# Extract data for class 0
C0 = X[:31,0:2]
print(" Data for class 0")
print(" ")
print(C0)
```

```python
# Extract data for class 1
C1 = X[31:,0:2]
print(" Data for class 1")
print(" ")
print(C1)
```

```python
# Now finding mean for class 0
M0 = np.mean(C0)
print("Mean of data  for class 0 is")
print(" ")
print(M0)
```

```python
# Now finding mean for class 1
M1 = np.mean(C1)
print("Mean of data  for class 1 is")
print(" ")
print(M1)
```

```python
# Now finding  terms for scatter matrix SW
c0 = np.array(C0)
first = []
for i in range(len(c0)):
    first.append(c0[i] - M0)

First = np.array(first)
    
print(First)
```

```python
# taking transpose
First_trans = First.T
print(First_trans)
```

```python
# Now finding first term
first_term = First_trans.dot(First)
print(first_term)
```

```python
c1 = np.array(C1)
second = []
for i in range(len(c1)):
    second.append(c1[i] - M1)

Second = np.array(second)
    
print(Second)
```

```python
# taking transpose
Second_trans = Second.T
print(Second_trans)
```

```python
# Now finding second term
second_term = Second_trans.dot(Second)
print(second_term)
```

```python
#Now finding scatter matrix SW
print("Scatter matrix is ")
print(" ")
SW = first_term + second_term
print(SW)
```

```python
# Now finding inverse of scatter matrix SW
SW_inver = np.linalg.inv(SW)
print(" Inverse of scatter matrix is  ")
print(" ")
print(SW_inver)
```

```python
# Now finding best direction that is W 
W = SW_inver*(M1-M0)
print(" Best direction W is ")
print(" ")
print(W)
```
